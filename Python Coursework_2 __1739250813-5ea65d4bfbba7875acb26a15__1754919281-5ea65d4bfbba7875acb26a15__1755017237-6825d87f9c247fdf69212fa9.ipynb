**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# code here
x=list(range(101))
y=[i*2 for i in x]
z=[i**2 for i in x]
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('Main plot')

ax2=fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

plt.show()
```

**Now plot (x,y) and (x,z) on the axes.**

```python
# code here
import numpy as np
x=np.linspace(0,100,100)
y=2*x
z=x**2

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot (x, y) with blue dashed line
axs[0].plot(x, y, 'b--')
axs[0].set_title('(x, y)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Plot (x, z) with red solid line
axs[1].plot(x, z, 'r-')
axs[1].set_title('(x, z)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# code here
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins").dropna()
features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
corr_matrix = penguins[features].corr()

sns.heatmap(
    corr_matrix,
    annot=True,           
    cmap="inferno",       
    vmin=-1, vmax=1,     
    linewidths=0.5,       
    square=True           
)

# Add title
plt.title("Correlation Heatmap of Penguin Measurements")
plt.show()
```

```python
# code here
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean the penguins dataset
penguins = sns.load_dataset("penguins").dropna()

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="deep"  # Matches: Adelie=blue, Chinstrap=orange, Gentoo=green
)

# Add title and labels
plt.title("Bill Length vs Bill Depth")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")
plt.grid(True)
plt.show()
```
